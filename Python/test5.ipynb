```python
users = {}
events = {}
history = []
chat = {}
delimeter = '\n'

def addUser(nick, name, tel_number):
    if nick in users:
        return f'{nick} already exists in the database'
    else:
        users[nick] = {"name": name, "tel_number": tel_number}
        return f'{nick} was successfully added'
    
def addToChat(sender, receiver, event_id):
    if sender in chat:
        if receiver in chat[sender]:
            chat[sender][receiver].append(event_id)
        else:
            chat[sender][receiver] = [event_id]
    else:
        chat[sender] = {receiver:[event_id]}

def addEvent(sender, receiver, message):
    if sender not in users:
        return f'{sender} is not in user base'
    if receiver not in users:
        return f'{receiver} is not in user base'
    event_id = len(history)
    events[event_id] = {"sender":sender, "receiver":receiver, "message":message}
    history.append(event_id)
    addToChat(sender, receiver, event_id)
    addToChat(receiver, sender, event_id)
    print(chat)
    return f'ADDED : {sender}->{receiver} : {message}'


def getHistory():
    output = 'HISTORY' + delimeter + delimeter
    for event_id in history:
        sender_id = events[event_id]['sender']
        receiver_id = events[event_id]['receiver']
        message = events[event_id]['message']
        output += users[sender_id]['name'] + ' -> ' + users[receiver_id]['name']  + ' : ' + message + delimeter 
    return output

def getMessages(user):
    output = user + delimeter + 'MESSSAGES' + delimeter
    for _, msgs in chat[user].items():
        for event_id in msgs:
            sender_id = events[event_id]['sender']
            receiver_id = events[event_id]['receiver']
            message = events[event_id]['message']
            output += users[sender_id]['name'] + ' -> ' + users[receiver_id]['name']  + ' : ' + message + delimeter 
    return output


while True:
    command = input('command (new/event/history/get) ? >')

    if command == 'new':
        nick, name, tel = input('Nick, name, tel ? >').split(',')
        print(addUser(nick, name, tel))
    elif command == 'event':
        sender, receiver, message = input('Who, whom, message ? >').split(',')
        addEvent(sender, receiver, message)
    elif command == 'history':
        print(getHistory())
    elif command == 'get':
        user = input('what is username ? >')
        print(getMessages(user))
    else:
        print(f'Unknown command -> {command}')
        break
```

```python
from csv import reader
from time import time

sales = reader(open('sales.csv'))
next(sales)
sales = list(sales)

pur = reader(open('purchase.csv'))
next(pur)
pur = list(pur)

cli = reader(open('clients.csv'))
next(cli)
cli = list(cli)
client = {}
for id,name,num,adr in cli:
    client[id] = {'name':name, 'num':num, 'adress':adr}

dist = reader(open("distributor.csv"))
next(dist)
dist = list(dist)
distributor = {}
for id,name,num in dist:
    distributor[id] = {'name':name, 'num':num}
    
tov = reader(open('products.csv'))
next(tov)
tov = list(tov)
products = {}
for id, pro, weight in tov:
    products[id] = {'name':pro,'weight':weight}
    
# 1.Напишите программу на Python, которая рассчитает общее количество каждого товара, купленного у дистрибьюторов, 
# используя данные из файлов items.csv и purchase.csv.
def dist_bought():
    how = {}
    text = "BOUGHT\n"
    for row in pur:
        name = products[row[2]]['name']
        how[name] = how.get(name, 0) + int(row[3])
    for i in how:
        text+=f"{i}: {how[i]}\n"
    return text
# print(dist_bought())

# 2.Объедините файлы sales.csv и items.csv, чтобы рассчитать общий доход, полученный от продажи каждого товара в магазине.
def dohod():
    doh = {}
    text = "EARNINGS\n"
    for row in sales:
        name = products[row[2]]['name']
        doh[name] = doh.get(name, 0) + float(row[5])
    for i in doh:
        text+=f"{i}: {doh[i]}\n"
    return text
# print(dohod())

# 3. Используя данные из purchase.csv, distributor.csv и items.csv, рассчитайте общую сумму, потраченную каждым 
# дистрибьютором на каждый товар, и определите, какой дистрибьютор поставил самый дорогой товар.
def obsh():
    mx = float('-inf')
    mn = None
    item = None
    dis = {}
    text = "DISTRIBUTORS:\n"
    for row in pur:
        name = distributor[row[1]]['name']
        tov = products[row[2]]['name']
        dis[name] = dis.get(name, {})
        dis[name][tov] = dis.get(tov, 0) + float(row[5])
        if float(row[4])>mx:
            mx = float(row[4])
            mn = name
            item = tov
    for i, x in dis.items():
        text+= f"\n{i} items:\n"
        for u in x:
            text+=f"{u}: {x[u]}\n"
    text+=f"\n\nMax priced item: {item} - {mx}\nDistributor: {mn}"
    return text
print(obsh())        

# 4. Объедините sales.csv и clients.csv, чтобы выяснить, какой клиент совершил наибольшую сумму покупок в магазине.
def most_bought():
    b = {}
    for row in sales:
        name = client[row[1]]['name']
        b[name] = b.get(name, 0) + float(row[5])
    mx = max(b, key=b.get)
    return f"{mx} bought spent the most: {b[mx]}"
# print(most_bought())

# 5.Используя файлы sales.csv, clients.csv и items.csv, найдите топ-5 клиентов по количеству разных товаров, которые они приобрели.
def most_different():
    f = {}
    mx = float('-inf')
    nm = None
    for row in sales:
        name = client[row[1]]['name']
        f[name] = f.get(name, [])
        if row[2] not in f[name]:
            f[name].append(row[2])
    # for i in f:
    #     if len(v)<=4:
    #         v[i] = f[i]
    #     else:
    #         min_key = min(v, key=lambda x: len(v[x]))
    #         if len(v[min_key])<len(f[i]):
    #             v.pop(min_key)
    #             v[i] = f[i]
    v = dict(sorted(f.items(), key = lambda item: len(item[1]), reverse =True)[:5])
    text = ""
    for i in v:
        text+= f"{i}: {len(v[i])}\n"
    return text
# print(most_different())

# 6. Объедините purchase.csv, items.csv и distributor.csv, чтобы рассчитать среднюю цену, 
# заплаченную за каждый товар магазином у разных дистрибьюторов.
def avg_tov():
    avg = {}
    avg1 = {}
    for row in pur:
        name = products[row[2]]['name']
        avg[name] = avg.get(name, {})
        avg[name][row[1]] = float(row[4])
    for i in avg:
        avg1[i] = sum(avg[i].values())/len(avg[i])
    text = "AVERAGE:\n"
    for i in avg1:
        text+=f"{i}: {avg1[i]}\n"
    return text
# print(avg_tov())

# 7.Напишите программу на Python, чтобы определить, какой товар приносит наибольшую прибыль,
# сравнив стоимость покупки (из purchase.csv) и цену продажи (из sales.csv).
def most_pribil():
    sa = {}
    sa2 ={}
    mx = float('-inf')
    nm = None
    for row in sales:
        name = products[row[2]]['name']
        sa[name] = float(row[4])
    for row in pur:
        name = products[row[2]]['name']
        sa2[name] = sa[name] - float(row[4])
        if sa2[name]>mx:
            mx = sa2[name]
            nm = name

    return f"Max pribil item: {nm} - {mx}"
# print(most_pribil())

# 8. Объедините purchase.csv и sales.csv, чтобы проанализировать, продал ли магазин больше товаров,
# чем приобрел, за определенный период времени.
def b_or_s():
    c = 0
    c1 = 0
    for row in sales:
        c+=float(row[5])//float(row[4])
    for row in pur:
        c1+=int(row[3])
    return f"{c}\n{c1}"
# print(b_or_s())  

# 9. Используя clients.csv, sales.csv и items.csv, напишите программу, чтобы определить клиента, 
# который купил наибольшее количество определенного товара.
def most():
    c ={}
    mx = float('-inf')
    mn = None
    mnn = None
    for row in sales:
        name = client[row[1]]['name']
        c[name] = c.get(name, {}) 
        c[name][products[row[2]]['name']] = c[name].get(products[row[2]]['name'], 0)+ float(row[3])
    for i,x in c.items():
        for u in x:
            if x[u]>mx:
                mx = x[u]
                mn = u
                mnn = i
    return f"{mnn} bought the most {mn} - {mx}"
# print(most())

# 10.Напишите программу на Python, которая использует данные из distributor.csv и purchase.csv, 
# чтобы узнать, какой дистрибьютор потавил больше всего разных товаров в магазин.
def dist_dif():
    d = {}
    #for solution 2
    # mx = float('-inf')
    # mn = None
    for row in pur:
        name = distributor[row[1]]['name']
        d[name] = d.get(name, set())
        d[name].add(row[2])
    #solution 1
    mn = max(d, key=lambda k: len(d[k]))
    return f"{mn} --> {len(d[mn])}"
    #solution 2
    # for i in d:
    #     if len(d[i])>=mx:
    #         mx = len(d[i])
    #         mn = i
    # return f"{mn} --> {mx}"
# print(dist_dif())

# 11.Объедините sales.csv и items.csv, чтобы рассчитать среднюю цену продажи каждого товара в магазине.
def avg_():
    avg = {}
    avg1 = {}
    for row in sales:
        name = products[row[2]]['name']
        avg[name] = avg.get(name, {})
        avg[name][row[1]] = float(row[4])
    for i in avg:
        avg1[i] = sum(avg[i].values())/len(avg[i])
    text = "AVERAGE:\n"
    for i in avg1:
        text+=f"{i}: {avg1[i]}\n"
    return text
# print(avg_())


# 12.Напишите программу на Python, чтобы рассчитать общий доход от клиентов, 
# которые купили товары у определенного дистрибьютора, используя данные из sales.csv, clients.csv и distributor.csv.

# 13.Используя items.csv, purchase.csv и sales.csv, выясните, какой товар имеет
# наибольший товарооборот (т.е. наиболее часто покупаемый и продаваемый).
def tovoborot():
    d = {}
    text = 'TOV OBOROT:\n'
    for row in sales:
        name = products[row[2]]['name']
        d[name] = d.get(name, 0) + float(row[3])
    for i in d:
        text+=f"{i}: {d[i]}\n"
    return text
# print(tovoborot())


# 14.Объедините purchase.csv, distributor.csv и sales.csv, чтобы узнать, 
# какой доход от продаж был получен от товаров, поставляемых каждым дистрибьютором.

# 15.Используя clients.csv, sales.csv и items.csv, определите, какой товар чаще всего покупают постоянные клиенты.
def post_clients():
    d = {}
    d1 = {}
    for row in sales:
        name = client[row[1]]['name']
        d[name] = d.get(name, 0)+1
        d1[name] = d1.get(name, {})
        d1[name][products[row[2]]['name']] = d1[name].get(products[row[2]]['name'], 0) +1
    d = dict(sorted(d.items(), key=lambda x:x[1], reverse = True)[:10])
    text = "TOP CLIENTS MOST BOUGHT ITEMS:\n"
    d2 = {}
    for i in d:
        for x in d1[i]:
            d2[x] = d2.get(x, 0) + d1[i][x]
    return max(d2.items(), key=lambda x:x[1])
# print(post_clients())

# 16.Напишите программу на Python, чтобы рассчитать среднюю стоимость товаров, купленных у каждого дистрибьютора,
# и сравнить её со средней ценой продажи клиентам, используя данные из purchase.csv, distributor.csv и sales.csv.

# 17.Объедините sales.csv и purchase.csv, чтобы проанализировать общее количество товаров, 
# купленных у дистрибьюторов, по сравнению с общим количеством проданных товаров клиентам для каждого товара.
def all():
    d ={}
    d2 = {}
    for row in sales:
        name= products[row[2]]['name']
        d[name] = d.get(name, 0) + float(row[3])
    for row in pur:
        name = products[row[2]]['name']
        d2[name] = d.get(name, 0) + float(row[3])
    for i in d:
        if i in d2:
            d[i] -= d2[i]
    return d
# print(all())

# 18.Напишите программу на Python, которая использует данные из distributor.csv, purchase.csv и sales.csv, 
# чтобы сгенерировать отчет о том, какие товары дистрибьютора приносят наибольший доход от продаж.

# 19. Используя данные clients.csv и sales.csv, напишите программу, чтобы определить, 
# какие клиенты принесли наибольший доход в магазин за определенный период времени.
def most_time(tim, tim2):
    d ={}
    for row in sales:
        if tim<=row[0]<=tim2:
            name = client[row[1]]['name']
            d[name] = d.get(name, 0) + float(row[5])
    return max(d.items(), key=lambda x:x[1])
# print(most_time('2023-06-23', '2023-07-25'))


# 20.Объедините sales.csv, items.csv и purchase.csv, чтобы выяснить, есть ли товары, 
# которые магазин продал в большем количестве, чем купил у дистрибьюторов.
def more_than():
    d ={}
    d2 = {}
    for row in sales:
        name= products[row[2]]['name']
        d[name] = d.get(name, 0) + float(row[3])
    for row in pur:
        name = products[row[2]]['name']
        d2[name] = d.get(name, 0) + float(row[3])
    for i in d:
        if i in d2:
            d[i] -= d2[i]
            if d[i]>-1:
                d.pop(i)
    return d
# print(more_than())
# 21.Используя clients.csv, sales.csv и items.csv, рассчитайте среднее количество каждого товара, 
# приобретенного клиентами при оптовых заказах (более определенного количества).
def optom():
    d = {}
    for row in sales:
        name = products[row[2]]['name']
        if float(row[3])>100:
            d[name] = d.get(name, {'summa':0, 'count':0})
            d[name]['summa'] += float(row[3])
            d[name]['count'] += 1
    d1 = {}
    for i in d:
        d1[i] = d[i]['summa']/d[i]['count']
    return d1
# print(optom())
# 22.Напишите программу на Python, которая использует данные из purchase.csv, items.csv и sales.csv,
# чтобы определить товары, которые часто покупаются у дистрибьюторов, но редко продаются клиентам.

# 23.Объедините distributor.csv, purchase.csv и sales.csv, чтобы сгенерировать отчет, 
# показывающий среднее время между покупкой товара у дистрибьютора и его продажей клиенту.

# 24.Используя clients.csv, sales.csv и items.csv, найдите самые популярные товары среди клиентов, 
# которые сделали покупки на сумму выше определенного порога.
def bought_clients():
    d = {}
    d1 = {}
    for row in sales:
        name = client[row[1]]['name']
        d[name] = d.get(name, 0 ) + float(row[5])
        d1[name] = d1.get(name, {})
        d1[name][products[row[2]]['name']] = d1[name].get(products[row[2]]['name'], 0) +1
    for i in d:
        if d[i]<100000:
            d.pop(i)
    d2 = {}
    for i in d:
        for x in d1[i]:
            d2[x] = d2.get(x, 0) + d1[i][x]
    return max(d2.items(), key=lambda x:x[1])
# print(bought_clients())
# 25.Напишите программу на Python, чтобы проанализировать, какую долю от общего дохода приносят товары,
# купленные у топ-3 дистрибьюторов, используя данные из distributor.csv, purchase.csv и sales.csv.

# 26.Объедините items.csv и sales.csv, чтобы рассчитать 
# общее количество проданных товаров и сравнить его с общим количеством товаров, имеющихся на складе.

# 27.Используя данные из distributor.csv, purchase.csv и sales.csv, выясните, 
# товары какого дистрибьютора приносят наибольшую среднюю цену при продаже клиентам.

# 28. Объедините clients.csv, sales.csv и items.csv, чтобы определить клиентов, 
# которые купили все товары из определенной категории или группы товаров.

# 29. Напишите программу на Python, которая объединяет данные из purchase.csv и sales.csv, 
# чтобы проанализировать маржу прибыли для каждого дистрибьютора, сравнивая стоимость покупки товаров с доходом от их продажи.

# 30.Используя clients.csv, sales.csv, purchase.csv и items.csv, создайте отчет, 
# который покажет клиента, купившего наибольшее разнообразие товаров в магазине.
def diff():
    d = {}
    for row in sales:
        name = client[row[1]]['name']
        d[name] = d.get(name, set())
        d[name].add(row[2])
    mn = max(d, key=lambda k: len(d[k]))
    return f"{mn} --> {len(d[mn])}"
# print(diff())
```
